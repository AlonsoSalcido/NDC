##  JSON to Product Data Frame

```python
# Create a Dartaframe
df = pd.DataFrame(columns=['product_ndc', 'generic_name', 'labeler_name', 'brand_name', 'finished', 'listing_expiration_date', 'marketing_category', 'dosage_form', 'product_type', 'marketing_start_date', 'product_id', 'brand_name_base', 'package_ndc', 'description', 'marketing_start_date', 'marketing_end_date', 'sample'])
```

## JSON to Ingredients Data Frame

```python
# Create a Dartaframe
ingredients_df = pd.DataFrame(columns=['product_ndc', 'name', 'strength'])
```

```python
rows = []
for product in data:
    product_ndc = product.get('product_ndc', None)
    active_ingredients = product.get('active_ingredients', [])
    for ingredients in active_ingredients:
        name = ingredients.get('name', None)
        strength = ingredients.get('strength', None)
        if name is not None and strength is not None:
            row = {'product_ndc': product_ndc, 'name': name, 'strength': strength}
            rows.append(row)

ingredients_df = pd.DataFrame(rows)
ingredients_df
```

```python
ingredients_df.shape
```

```python
# create a new dataframe with unique ids for each combination of 'name' and 'strength'
id_df = ingredients_df[['name', 'strength']].drop_duplicates().reset_index(drop=True).reset_index()
id_df.columns = ['ingredient_id', 'name', 'strength']

# merge the new dataframe with the original dataframe
ingredients_df = ingredients_df.merge(id_df, on=['name', 'strength'])

# print the result
print(ingredients_df)
```
